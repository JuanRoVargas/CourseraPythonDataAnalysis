# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Leer la tabla de interés
query = "SELECT * FROM master"
df = pd.read_sql_query(query, conn)

# Cerrar la conexión
conn.close()

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna())
plt.title('Histograma de Compensación Convertida')
plt.xlabel('Compensación Convertida')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
if 'Age' in df.columns:
    plt.figure(figsize=(10, 6))
    plt.boxplot(df['Age'].dropna(), vert=False, patch_artist=True)
    plt.title('Diagrama de Caja de la Edad')
    plt.xlabel('Edad')
    plt.grid(True)
    plt.show()
else:
    print("La columna 'Age' no se encuentra en la tabla.")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
if 'Age' in df.columns and 'WorkWeekHrs' in df.columns:
    # Eliminar filas con valores nulos en ambas columnas
    df_filtered = df[['Age', 'WorkWeekHrs']].dropna()
    
    # Crear el diagrama de dispersión
    plt.figure(figsize=(10, 6))
    plt.scatter(df_filtered['Age'], df_filtered['WorkWeekHrs'], alpha=0.5)
    plt.title('Diagrama de Dispersión: Edad vs. Horas de Trabajo por Semana')
    plt.xlabel('Edad')
    plt.ylabel('Horas de Trabajo por Semana')
    plt.grid(True)
    plt.show()
else:
    print("Una o ambas columnas ('Age' y 'WorkWeekHrs') no se encuentran en la tabla.")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
if 'WorkWeekHrs' in df.columns and 'CodeRevHrs' in df.columns and 'Age' in df.columns:
    # Eliminar filas con valores nulos en las columnas relevantes
    df_filtered = df[['WorkWeekHrs', 'CodeRevHrs', 'Age']].dropna()
    
    # Crear el diagrama de burbujas
    plt.figure(figsize=(12, 8))
    plt.scatter(
        df_filtered['WorkWeekHrs'], 
        df_filtered['CodeRevHrs'], 
        s=df_filtered['Age'] * 10,  # Ajusta el tamaño de las burbujas multiplicando por un factor
        alpha=0.5,
        edgecolors='w', 
        cmap='viridis'
    )
    plt.title('Diagrama de Burbujas: Horas de Trabajo vs. Horas de Revisión de Código')
    plt.xlabel('Horas de Trabajo por Semana')
    plt.ylabel('Horas de Revisión de Código')
    plt.grid(True)
    plt.colorbar(label='Edad')  # Muestra una barra de colores para la variable de tamaño
    plt.show()
else:
    print("Una o más columnas necesarias no se encuentran en la tabla.")
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Leer la tabla de interés
query = "SELECT * FROM master"
df = pd.read_sql_query(query, conn)

database_counts = df['OpenSource'].value_counts()

# Seleccionar las 5 principales bases de datos
top_5_databases = database_counts.head(5)

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_databases,
    labels=top_5_databases.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired(range(len(top_5_databases)))
)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Leer la tabla de interés
query = "SELECT * FROM master"
df = pd.read_sql_query(query, conn)

filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calcula la mediana de WorkWeekHrs y CodeRevHrs
median_workweekhrs = filtered_df['WorkWeekHrs'].median()
median_coderevhrs = filtered_df['CodeRevHrs'].median()

# Prepara los datos para el gráfico
median_data = pd.DataFrame({
    'Metric': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median Hours': [median_workweekhrs, median_coderevhrs]
})

# Crea el gráfico apilado
fig, ax = plt.subplots()
ax.bar(median_data['Metric'], median_data['Median Hours'], color=['#1f77b4', '#ff7f0e'])

ax.set_ylabel('Median Hours')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Leer la tabla de interés
query = "SELECT * FROM master"
df = pd.read_sql_query(query, conn)

mainbranch_counts = df['MainBranch'].value_counts()

# Crear el gráfico de barras horizontales
plt.figure(figsize=(10, 7))
mainbranch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branches')
plt.show()
```
