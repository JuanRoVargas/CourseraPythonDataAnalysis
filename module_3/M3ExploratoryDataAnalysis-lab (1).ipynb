## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV en un DataFrame
df = pd.read_csv("m2_survey_data.csv")

# Asegurarse de que la columna 'ConvertedComp' no tenga valores nulos
df = df.dropna(subset=['ConvertedComp'])

# Configurar el estilo del gráfico
plt.style.use('seaborn-whitegrid')

# Crear el gráfico de distribución
plt.figure(figsize=(12, 6))
plt.hist(df['ConvertedComp'], bins=30, color='blue', edgecolor='black', alpha=0.7)

# Agregar etiquetas y título
plt.xlabel('Salario Anual Convertido (USD)')
plt.ylabel('Frecuencia')
plt.title('Distribución del Salario Anual Convertido')

# Mostrar el gráfico
plt.show()
```

Plot the histogram for the column `ConvertedComp`.


```python
plt.figure(figsize=(10, 6))  # Opcional: tamaño de la figura
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='black')  # Crear el histograma
plt.title('Distribución de los Salarios Convertidos')  # Título del gráfico
plt.xlabel('Salario Convertido')  # Etiqueta del eje x
plt.ylabel('Frecuencia')  # Etiqueta del eje y
plt.show()
```

What is the median of the column `ConvertedComp`?


```python
median_converted_comp = df['ConvertedComp'].median()
print("La mediana de la columna ConvertedComp es:", median_converted_comp)
```

How many responders identified themselves only as a **Man**?


```python
count_men = df[df['Gender'] == 'Man'].shape[0]
print("El número de encuestados que se identificaron únicamente como 'Man' es:", count_men)
```

Find out the  median ConvertedComp of responders identified themselves only as a **Woman**?


```python
df_women = df[df['Gender'] == 'Woman']
median_converted_comp_women = df_women['ConvertedComp'].median()
print("La mediana de la columna ConvertedComp para las encuestadas que se identificaron como 'Woman' es:", median_converted_comp_women)
```

Give the five number summary for the column `Age`?

```python
min_age = df['Age'].min()
q1_age = df['Age'].quantile(0.25)
median_age = df['Age'].median()
q3_age = df['Age'].quantile(0.75)
max_age = df['Age'].max()

print("Resumen de cinco números para la columna Age:")
print("Mínimo:", min_age)
print("Primer cuartil (Q1):", q1_age)
print("Mediana:", median_age)
print("Tercer cuartil (Q3):", q3_age)
print("Máximo:", max_age)
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
plt.figure(figsize=(10, 6))  # Opcional: tamaño de la figura
plt.boxplot(df['ConvertedComp'].dropna(), vert=False)  # Crear el box plot horizontal
plt.title('Box Plot de los Salarios Convertidos')  # Título del gráfico
plt.xlabel('Salario Convertido')  # Etiqueta del eje x
plt.show()  # Mostrar el gráfico
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)

iqr = q3 - q1

print("Primer Cuartil (Q1):", q1)
print("Tercer Cuartil (Q3):", q3)
print("Rango Intercuartílico (IQR):", iqr)
```

Find out the upper and lower bounds.


```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print("Límite Inferior:", lower_bound)
print("Límite Superior:", upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]
num_outliers = len(outliers)
print("Número de valores atípicos en la columna ConvertedComp:", num_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
df_no_outliers.head()
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
numeric_df = df.select_dtypes(include='number')

correlation_matrix = numeric_df.corr()

age_correlation = correlation_matrix['Age']
print(age_correlation)
```
